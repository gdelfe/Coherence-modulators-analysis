```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def plot_histo(marks,title,labels):
        
    fig, ax = plt.subplots(1, 1)
    ax.hist(marks, bins=np.arange(len(labels)+1),alpha=0.7)
    
    plt.grid(color='white', lw = 0.5, axis='x')
    plt.title(title, loc = 'left', fontsize = 14)
    plt.xlabel('\nSender-Receiver area', fontsize=14)
    plt.ylabel('# of modulators', fontsize=14)

    rects = ax.patches
    max_h = 0
    for rect, label in zip(rects, labels):
        height = rect.get_height()
        max_h = max(max_h,height)
        ax.text(rect.get_x() + rect.get_width() / 2, height+0.02, label,
                ha='center', va='bottom',fontsize=9)

    axes = plt.gca()
    axes.set_ylim([0,max_h + 2])
    values = np.arange(0,max_h+ 3,2)
    plt.yticks(values)
    
    return fig
```

```python
# Creating dataset
monkey = 'Maverick'
freq_band = 'theta_band'

CN_M1 = np.full(32,0)
CN_CN = np.full(31,1)
OFC_CN = np.full(21,2)
CN_ACC = np.full(9,3)
OFC_M1 = np.full(6,4)
CN_dPFC = np.full(4,5)
OFC_ACC = np.full(3,6)
labels = ['CN-M1','CN-CN','OFC-CN','CN-ACC','OFC-M1','CN-dPFC','OFC-ACC']
title = 'Maverick - Theta modulators for each SR area'

marks = np.concatenate((OFC_CN,OFC_M1,CN_CN,CN_M1,CN_ACC,CN_dPFC,OFC_ACC),axis=0)
```

```python
path = '/mnt/pesaranlab/People/Gino/Coherence_modulator_analysis/Shaoyu_data/{}/Resting_State/{}/histograms'.format(monkey,freq_band)
fig = plot_histo(marks,title,labels)
fig.savefig('Subplot_ex1.png')
fig.savefig('{}/{}_histogram_{}_by_SR_area.png'.format(path,monkey,freq_band)) 
plt.show()
```
